```python
def loan_count(df):
    n = len(df['LOAN_SEQUENCE_NUMBER'].unique())
    return n
 
```

SECOND COLUMN

```python
def paid_loan(df):

    x = len(df[df['ZERO_BALANCE_CODE'] == 1])
    y = len(df[df['ZERO_BALANCE_CODE'] == 6])
    z = x+y
    Total = (z/len(df['LOAN_SEQUENCE_NUMBER'].unique()))*100
    return Total
```

THIRD COLUMN

```python
def defaulted_loan(df):
    x = len(df[df['ZERO_BALANCE_CODE'] == 9])
    y = len(df[df['ZERO_BALANCE_CODE'] == 3])
    z = x+y
    Total = (z/len(df['LOAN_SEQUENCE_NUMBER'].unique()))*100
    return Total
```

FOURTH COLUMN

```python
def remaining_loans(df):
    m= len(df['LOAN_SEQUENCE_NUMBER'].unique())-len(df[df['ZERO_BALANCE_CODE'] == 1])-len(df[df['ZERO_BALANCE_CODE'] == 6])-len(df[df['ZERO_BALANCE_CODE'] == 9])-len(df[df['ZERO_BALANCE_CODE'] == 3])
    Total = (m/len(df['LOAN_SEQUENCE_NUMBER'].unique()))*100
    return Total
```

```python
def loan_180(df):
    
    df = df.convert_objects(convert_numeric=True)
    df['CURRENT_LOAN_DELINQUENCY_STATUS'].replace('nan',9999, inplace = True)
    df['CURRENT_LOAN_DELINQUENCY_CORRECTED'] = df['CURRENT_LOAN_DELINQUENCY_STATUS'].replace('nan',9999)
    df['CURRENT_LOAN_DELINQUENCY_CORRECTED'] = df['CURRENT_LOAN_DELINQUENCY_STATUS'].replace('R', 9999)
    columns =['LOAN_SEQUENCE_NUMBER','CURRENT_LOAN_DELINQUENCY_CORRECTED']
    df1 =pd.DataFrame(df, columns=columns)
    
    df1 = df1.drop_duplicates()
    n = len(df1[(df1['CURRENT_LOAN_DELINQUENCY_CORRECTED'].astype(int) == 6) ])

    Total = (n/len(df['LOAN_SEQUENCE_NUMBER'].unique())*100)
    return Total
```

```python
def loan_R(df):
    df = df.convert_objects(convert_numeric=True)
    df['CURRENT_LOAN_DELINQUENCY_STATUS'].replace('nan',9999, inplace = True)
    df['CURRENT_LOAN_DELINQUENCY_CORRECTED'] = df['CURRENT_LOAN_DELINQUENCY_STATUS'].replace('R', 9999)
    columns =['LOAN_SEQUENCE_NUMBER','CURRENT_LOAN_DELINQUENCY_CORRECTED']
    df1 =pd.DataFrame(df, columns=columns)
    
    df1 = df1.drop_duplicates()
    n = len(df1[(df1['CURRENT_LOAN_DELINQUENCY_CORRECTED'].astype(int) == 9999) ])

    Total = (n/len(df['LOAN_SEQUENCE_NUMBER'].unique())*100)
    return Total
```

ELEVENTH COLUMN

```python
def modified_loan(df):
    columns =['LOAN_SEQUENCE_NUMBER','MODIFICATION_FLAG']
    df2 =pd.DataFrame(df, columns=columns)
    
    df2 = df2.drop_duplicates()

    n = len(df2[(df2['MODIFICATION_FLAG'] == 'Y') ])

    Total = (n/len(df['LOAN_SEQUENCE_NUMBER'].unique()))*100
    return Total
```

```python
def nonprepayment_Loss(df):
    df3 = pd.DataFrame()
    df3 = df[(df['REPURCHASE_FLAG'] == 'Y') & (df['ZERO_BALANCE_CODE'] == 6)]



    n = df3['Actual_Loss_Calculation'].sum()
    return n
```

THIRTEEN

```python
def Total_loss(df):
    n = df['Actual_Loss_Calculation'].sum()
    return n
```

```python
def Total_Expenses(df):
    n = df['EXPENSES'].sum()
    return n
```

```python
def aged_loan(df):
    n = len(df[(df['LOAN_AGE'] > 60) ])
    Total = (n/len(df['LOAN_SEQUENCE_NUMBER']))*100
    return Total
```

SIXTEEN

```python
def matured_loan(df):
    n = len(df[(df['REMAINING_MONTHS_TO_LEGAL_MATURITY'] == 0 ) ])
    return n
```

```python
def loans_defaulted_percentage(df):

    df['CURRENT_LOAN_DELINQUENCY_CORRECTED'] = df['CURRENT_LOAN_DELINQUENCY_STATUS'].replace('R', 9999)
    columns =['LOAN_SEQUENCE_NUMBER','CURRENT_LOAN_DELINQUENCY_CORRECTED']
    df1 =pd.DataFrame(df, columns=columns)
    df1.drop_duplicates('LOAN_SEQUENCE_NUMBER', take_last=True, inplace = True)

    n = len(df1[(df1['CURRENT_LOAN_DELINQUENCY_CORRECTED'].astype(int) > 2) ])

    Total = n/50000
    Total = Total*100

    return Total
```

```python
def movingaverage(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')

x= dfb['Year']
y =dfb['Weighted Average Coupon (WAC)']


plt.plot(x,y,"k.")
y_av = movingaverage(y, 3)
plt.plot(x, y_av,"r")
plt.xticks(np.arange(min(x), max(x)+1, 2.0))
plt.xlabel("Year")
plt.ylabel("Weighted Average Coupon")
plt.grid(True)
plt.show()
```
